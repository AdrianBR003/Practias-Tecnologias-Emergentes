## Series

Las *Series* son el tipo de dato principal en Pandas. Una Serie es muy parecida a un array de Numpy. Sin embargo, una Serie puede tener etiquetas para cada eje, de forma que puede accederse por el nombre de etiqueta en lugar de por posición. Además pueden almacenar cualquier tipo de dato u objeto de Python, incluso dentro del mismo objeto Serie.

Para comenzar importamos la librería Pandas (junto con Numpy para realizar varios ejemplos).

```python
import numpy as np
import pandas as pd
```

### Creando Series

Se puede crear una Serie a partir de una lista, un array de Numpy o un diccionario. Para ello utilizamos el constructor *Series*. A continuación vamos a ver algunos ejemplos:

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
```

**A partir de listas**

Indicando como parámetro *data* la lista que vamos a transformar.

```python
pd.Series(data=d)
```

```python
pd.Series?
```

Tambien se puede pasar una lista de etiquetas para cada fila:

```python
pd.Series(data=my_list,index=labels)
```

```python
pd.Series(data=my_list)
```

También se pueden indicar la lista y las etiquetas de fila como primer y segundo parámetro (sin indicar el nombre):

```python
pd.Series(my_list,labels)
```

**A partir de arrays de NumPy**

```python
pd.Series(arr)
```

```python
pd.Series(arr,labels)
```

**A partir de un Diccionario**

```python
pd.Series(d)
```

### Tipos de datos de un objeto Series

Un objeto Series de Pandas puede almacenar cualquier tipo de dato u objeto. Desde cadenas de texto:

```python
pd.Series(data=['uno','dos','tres'])
```

Definiciones de funciones:

```python
pd.Series([sum,print,len])
```

```python
funciones=pd.Series([sum,print,len])
funciones[1]('hola mundo')
```

Hasta objetos de distintos tipos:

```python
pd.Series([1.7,'no valido',9,3j])
```

## Utilizando índices

La clave para utilizar un objeto Series está en comprender sus índices. Pandas puede utilizar como índices nombres o números para acceder a sus datos. Trabaja de la misma manera que una tabla hash o un diccionario.

Veamos algunos ejemplos:

```python
ser1 = pd.Series([1,2,3,4],index = ['USA', 'Germany','USSR', 'Japan'])                                   

ser1
```

```python
ser2 = pd.Series([1,2,5,4],index = ['USA', 'Germany','Italy', 'Japan'])                                   

ser2
```

```python
ser1['USA']
```

Cualquier operación que se realice estará basada en los índices de las Series:

```python
ser1 + ser2
```

## DataFrames

Los DataFrames de Pandas son las estructuras mas utilizadas y están inspiradas en el lenguaje de programación R. Un DataFrame es como un grupo de objetos de tipo Series que comparten un mismo índice, y en el que cada serie puede tener un nombre. En resumen, es una estructura de dos dimensiones donde cada fila y cada columna puede tener una etiqueta.

```python
df = pd.DataFrame(np.random.randn(5,4), index='A B C D E'.split(), columns='W X Y Z'.split())
df
```

### Selección e índices.

Veamos varias maneras de manipular un DataFrame. En primer lugar podemos acceder a toda una columna indicando el nombre de dicha columna:

```python
df['W']
```

```python
df['W']
```

```python
df['W']['A']
```

Si queremos obtener los datos de varias columnas, hay que utilizar una lista con el nombre de las columnas:

```python
df[['W','Z']]
```

Cuando obtenemos un columna lo que nos devuelve es un objeto de tipo Series:

```python
type(df['W'])
```

**Creando un nueva columna:**

```python
df['new'] = df['W'] + df['Y']#crea una nueva columna

df
```

**Eliminando columnas**: Con el método *drop* podemos obtener un DataFrame sin la columna indicada.

```python
df.drop('new',axis=1)# en realidad crea una copia, el dataframe sigue siendo el mismo
```

Sin embargo, el funcionamiento por defecto del método no elimina la columna del DataFrame.

```python
df
```

Para que elimine la columna del DataFrame hay que indicarselo de forma explícita con el parámetro *inplace*:

```python
df.drop('new',axis=1,inplace=True) #el parámetro inplace suelen tenerlo todos los métodos,
#permite no crear copias sino usar el mismo dataframe

df
```

También se pueden eliminar filas con este método:

```python
df.drop('E',axis=0)
```

**Seleccionando filas**.

El método `loc` nos permite obtener los datos de una fila:
(los devuelve como una Serie)

```python
df.loc['A']
```

Podemos seleccionar una fila por su posición en lugar de su etiqueta utilizando el método `iloc`:

```python
df.iloc[2]
```

**Seleccionando subconjuntos de filas y columnas**

Se pasa fila y columna:

```python
df.loc['B','Y']
```

Si se pasa lista de filas y lista de columnas:

```python
df.loc[['A','B'],['W','Y']]
```

```python
df
```

### Selección condicional

Una característica importante de Pandas es la selección condicional, muy parecida a la utilizada en Numpy.

```python
df
```

Si comparamos un DataFrame completo con un valor obtendremos un nuevo DataFrame con valores lógicos (True/False) indicando que valores del DataFrame cumple la condición.

```python
df>0
```

Este DataFrame de lógicos se puede utilizar para seleccionar aquellos valores que tienen su valor a verdadero (`True`):

```python
df[df>0]
```

```python
df[df>0]
nuevo=df.fillna(75)
nuevo
```

```python
df[df>0]=100
df
```

```python
df[df['W']>0]
df
```

Si utilizamos esta característica para una sóla columna, podemos filtrar el DataFrame para obtener sólo las filas cuya columna cumpla la condición dada:

```python
df[df['W']>0]
```

```python
df[df['W']>0]['Y']
```

```python
df[df['W']>0][['Y','X']]
```

Se pueden aplicar varias condiciones utilizando los operadores lógicos *O* (`|`) e *Y* (`&`) con paréntesis:

```python
df[(df['W']>0) & (df['Y'] < -1)]
```

### Mas detalles sobre los índices

Sobre los índices se pueden realizar distintos tipos de operaciones, como por ejemplo: reiniciarlos, cambiarlos por otros nuevos o crear una jerarquía de índices. Los índices son como otro tipo de datos

```python
df = pd.DataFrame(np.random.randn(5,4), index='A B C D E'.split(), columns='W X Y Z'.split())
df
```

**Resetear índices**: Utilizamos el método `reset_index` para inicializar los índices a sus valores numéricos por defecto (empezando por 0). Esto nos creará una nueva columna llamada `index` con los índices antíguos: 

```python
df.reset_index()
```

Pero no modifica el objeto, en realidad lo que hace es una copia.

```python
df
```

**Nuevos índices**: Para crear nuevos índices, primero tenemos que crear una nueva columna con los nombres de los índices y posteriormente indicamos que esta nueva columna se utilizará como índices:

```python
newind = 'CA NY WY OR CO'.split()

df['States'] = newind

df
```

```python
df.set_index('States')
```

Todas estas operaciones no se realizarán de forma permanente en el DataFrame a menos que se le indíque implícitamente con el parámetro `inplace`.

```python
df
```

```python
df.set_index('States',inplace=True)

df
```

### Multi-índices y jerarquía de índices.

Un DataFrame puede tener varias columnas índices.

A continuación vamos a ver como trabajar con multi-índices. Primero vamos a crear un pequeño ejemplo de un DataFrame que tiene varios índices (e incluso jerárquicos):

```python
# Creamos los multi-índices
outside = ['G1','G1','G1','G2','G2','G2']
inside = [1,2,3,1,2,3]
tuples = list(zip(outside,inside))
hier_index = pd.MultiIndex.from_tuples(tuples)

hier_index
```

```python
df = pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

Veamos algunos ejemplo de como funcionan estos índices. Podemos acceder al primer índice utilizando el método `loc` y obtendremos un *sub-dataframe*:

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

Los índices utilizados no tienen nombre propio, pero se les puede asignar un nombre a cada uno para diferenciarlos.

```python
df.index.names
```

```python
df.index.names = ['Group','Num']

df
```

Con el método `xs` se pueden obtener los datos de los índices que queramos:

```python
df.xs('G1')
```

```python
df.xs(['G1',1])
```

```python
df.xs(1,level='Num')
```

## Datos no válidos.

Cuando filtramos los datos de un DataFrame, los datos que no cumplen la condición se representan como `NaN`(*Not a Number*). Estos datos no datos también se conocen como *datos perdidos*.

Veamos algunos métodos que nos permiten trabajar con este tipo de datos en Pandas.

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})

df
```

**dropna**: Permite recuperar aquellos datos válidos del DataFrame. Quita filas que tenga algún valor NaN. No modifica dataset original

```python
df.dropna()
```

```python
df.dropna(axis=1) # Elimina columnas que tenga algún valor NaN
```

Con el parámetro `thresh` se puede indicar a partir de que cantidad de datos no válidos no se recuperan.

```python
df.dropna(thresh=2)
```

**fillna**: Se utiliza para indicar el valor por defecto por el que se sustituirá los datos no válidos.

```python
df.fillna(value='FILL VALUE')
```

```python
df.fillna(value=0)['A'].mean() #primero rellena un 0 y despues hace media (1+2+0)/3 que es 1
```

```python
df['A'].fillna(value=df['A'].mean())
# Los NaN los quita y pone la media de la columna (de los valores válidos)
```

## GroupBy

El método `groupby` permite agrupar datos por filas o columnas y utilizar funciones de agregación.

```python
data = {'Compañia':['GOOGLE','GOOGLE','MICROSOFT','MICROSOFT','FACEBOOK','FACEBOOK'],
       'Persona':['Juan','Miguel','Maria','Vanesa','Carlos','Sara'],
       'Sueldo':[200,120,340,124,243,350]}

df = pd.DataFrame(data)

df
```

Ahora podemos usar el método `groupby` para agrupar filas a partir del nombre de la columna. Por ejemplo, podemos agrupar por *Compañia*. Esto creará un objeto DataFrameGroupBy:

```python
df.groupby('Compañia')
```

Si guardamos el objeto en una variable, podemos llamar a diversos métodos de agregación y ver sus resultados:

```python
by_comp = df.groupby("Compañia")
#Media
by_comp.mean()
```

```python
#Desviación estándar
by_comp.std()
```

```python
#Mínimo
by_comp.min()
```

```python
#Máximo
by_comp.max()
```

```python
#Contar
by_comp.count()
```

También se puede sacar diversos valores estadísticos:

```python
#Estadísticas
by_comp.describe()
```

### Concatenación

La concatenación básicamente une los DataFrames. Hay que tener en cuenta que las dimensiones deben coindicir en el eje en que se concatena. Para la concatenación utilizamos `pd.concat` pasándole una lista con los DataFrames que queremos concatenar:

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[0, 1, 2, 3]) 

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[0, 1, 2, 3])
pd.concat([df1,df2,df3],axis=1)
```

### Mezcla

La función `merge` nos permite mezclar DataFrames utilizando una columna como clave de mezcla Dicha mezcla la realiza de forma similar a como lo hace el `JOIN` de `SQL`. Vamos a crear un par de DataFrames de ejemplo para ilustrar la mezcla:

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'keyA': ['A0', 'A1', 'A2', 'A3'],
                     'keyB': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K2', 'K1', 'K3'],
                          'keyC': ['C0', 'C1', 'C2', 'C3'],
                          'keyD': ['D0', 'D1', 'D2', 'D3']})    
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='key')
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'keyA': ['A0', 'A1', 'A2', 'A3'],
                     'keyB': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K4', 'K1', 'K3'],
                          'keyC': ['C0', 'C1', 'C2', 'C3'],
                          'keyD': ['D0', 'D1', 'D2', 'D3']})  
pd.merge(left,right,how='inner',on='key')
```

```python
pd.merge(left,right,how='outer',on='key')
```

```python
pd.merge(left,right,how='left',on='key')
```

```python
pd.merge(left,right,how='right',on='key')
```

Otro ejemplo mas complejo sería:

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
#Mezclar sólo claves comunes
pd.merge(left, right, on=['key1', 'key2'])
```

```python
#Mezclar utilizando todas las claves
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
#Mezclar utilizado claves del segundo DataFrame
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
#Mezclar utilizando claves del primer DataFrame
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

### Unión

El método `join` realiza una operación similar a la mezcla, solo que utiliza los propios índices del DataFrame, en lugar de las columnas indicadas.

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])

left.join(right)
```

```python
#Utiliza la unión de los índices de ambos DataFrames
left.join(right, how='outer')
```

### Resumen de datos.

Un resumen de datos es similar a las tablas dinámicas de Excel o las *tablas pivote* de OpenOffice.

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
        'B':['one','one','two','two','one','one'],
        'C':['x','y','x','y','x','y'],
        'D':[1,3,2,5,4,1]}

df = pd.DataFrame(data)

df
```

```python
df.pivot_table(values='D',index=['A', 'B'],columns=['C'])
```
