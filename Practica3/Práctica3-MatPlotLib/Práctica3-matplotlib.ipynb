## Introduccion.

Matplotlib es una excelente librería para realizar gráficas en 2D y 3D. Sus principales características son:

* Es fácil empezar a trabajar con ella, realizando gráficas sencillas.
* Permite etiquetas y textos en formato $\LaTeX$.
* Se puede controlar perfectamente cualquier elemento de una figura, incluyendo el tamaño y la resolución.
* Genera imágenes de gran calidad en varios formatos, como por ejemplo PNG, PDF, SVG o EPS.
* Se pueden realizar gráficas interactivas.
* Permite controlar todas las características de una figura como si se tratara de un programa. De esta forma, se pueden reproducir fácilmente todos los pasos para generar el mismo tipo de figura cuando cambian los datos.

Puedes ver mas información sobre Matplotlib en: http://matplotlib.org/

Existen diversas formas para empezar a trabajar con Matplotlib en Python. La mas mas sencilla sería importar todas las clases y métidos del módulo `pylab`:

```python

```

Aunque la forma mas ordenada sería importar el módulo `matplotlib.pyplot` bajo el alias `plt`:

````python
import matplotlib
import matplotlib.pyplot as plt
````

### Un primer ejemplo sencillo.

Vamos a realizar una primera gráfica, con el uso mas sencillo de Matplotlib. Para ello creamos dos vectores, uno que represente los valores del eje X, y otro con los datos de una función dada (en este caso $x^2$) que se representarán en el eje Y.

````python
import numpy as np
x = np.linspace(0, 5, 10)
y = x ** 2
````

A continuación inicializamos una figura simple, y le indicamos los datos que debe utilizar y los títulos y etiquetas que debe colocar en la figura.

````python
plt.figure()
plt.plot(x, y, 'r') #La 'r' indica que la linea será roja (red)
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.show()
````

Para dibujar varias gráficas en una podemos utilizar la función `subplot`, indicando el número de gráficas y su orden por filas y columnas.

````python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--') #Linea roja de rayas separadas
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-') #Línea verde (green) con asteriscos
plt.show()
````

## La API orientada a objetos de Matplotlib

La idea principal de este estilo es definir un objeto que represente la gráfica/figura, y aplicar todo tipo de funciones y acciones sobre ella. De esta forma, podemos manipular de forma clara mas de una figura o varias subgráficas.

Vamos a realizar el mismo ejemplo anterior, pero utilizando la API orientada a objetos.

````python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # izq., abajo, anchura, altura (rango 0 a 1)

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

Este ejemplo tiene algo más de código. Se incluye una línea con una función (`add_axes`) que nos permite definir la caja donde se representará la figura. La principal ventaja de este método consiste en que tenemos pleno control sobre donde se coloca cada gráfica en nuestra figura.

De esta forma podemos añadir más gráficas a la figura:

````python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #gráfica principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #gráfica interior

#Gráfica principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

#Gráfica interior
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig.tight_layout()
````

### Tamaño y resolución de la figura

Matplotlib permite especificar el tamaño, aspecto y resolución cuando se crea un objeto `Figure`. Para ello se utilizan los argumentos `figsize` y `dpi`.

El argumento `figsize` indica el tamaño y debe ser una tupla con el ancho y alto de la figura en pulgadas (*inches*).
El argumento `dpi` es la resolución, en puntos por pulgada (*dots-per-inch* o *pixel per inch*).

Estos parámetros se pueden utilizar tanto en la función `figure` como en la función `subplots`. 

Vamos crear una figura de 800x400 pixel y 100 puntos por pulgada:

````python
fig, axes = plt.subplots(figsize=(8,4), dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
````

### Leyendas.

Algunos valores de ejemplo son:

```python
ax.legend(loc=0) # Matplotlib elige el mejor lugar
ax.legend(loc=1) # Esquina superior derecha
ax.legend(loc=2) # Esquina superior izquierda
ax.legend(loc=3) # Esquina inferior izquierda
```

### Dando formato al texto.

Matplotlib también nos permite cambiar las fuentes de texto utilizadas, el tamaño del texto o incluso introducir texto en formato LaTeX.

Para utilizar el formato LaTeX en cualquier texto, sólo tenemos que enmarcarlo entre dos símbolos de dolar. Por ejemplo el texto `"$y=x^3$"` se mostraría como $y=x^3$.

Sin embargo, existe un pequeño problema con el formato LaTeX en Python. En LaTeX, se utiliza la barra invertida en algunas ordenes, por ejemplo `\alpha` para generar el símbolo $\alpha$. Pero la barra invertida en cadenas de texto de Python tienen otro significado, se utilizan como caracter de escape. Por lo tanto, para poder utilizar esta barra en formato LaTeX dentro de una cadena de Python, tenemos que indicar a Python que no interprete la cadena. Para ello utilizados *"cadenas en crudo"* (*raw strings*). Estas se indican precediéndolas con la letra `r`, como por ejemplo `r"\alpha"` o `r'\alpha'`:

````python
fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$\alpha$', fontsize=18)
ax.set_ylabel(r'$y$', fontsize=18)
ax.set_title('title');
````

Otro aspecto que podemos modificar en los textos de las gráficas son el tamaño y la fuente utilizada.

Para cambiar estos parámetros en todos los textos de las gráficas, modificamos de forma global los parámetros de Matplotlib:

````python
#Actualizar los parámetros de configuración de matplotlib
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'serif'})

fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2) 
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title');
````

Si la figura va a ir incorporada a una artículo en formato LaTeX, la fuente mas apropiada suele ser la *STIX*: 

````python
#Actualizar los parámetros de configuración de matplotlib
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})

fig, ax = plt.subplots()

ax.plot(x, x**2, label=r"$y = \alpha^2$")
ax.plot(x, x**3, label=r"$y = \alpha^3$")
ax.legend(loc=2)
ax.set_xlabel(r'$\alpha$')
ax.set_ylabel(r'$y$')
ax.set_title('title');
````

La siguiente orden configurará Matplotlib a su configuración original.

### Modificando el formato de las lineas

#### Colores

Matplotlib nos permite definir los colores de los elementos gráficos de diversas maneras. Podemos utilizar la sintaxis de MATLAB, donde `'b'` significa azul (*blue*), `'g'` significa verde (*green*), etc. También se puede utilizar la sintaxis de MATLAB para seleccionar el estilo de las líneas:

````python
#Estilo MATLAB
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-')  #línea azul con puntos
ax.plot(x, x**3, 'g--'); #línea verde discontinua
````

También se puede definir los colores con su código RGB en hexadecimal, o indicar de forma opcional un valor de transparencia.

Para ello utilizamos los argumentos `color` y `alpha`:

````python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="red", alpha=0.5) #Rojo semitransparente
ax.plot(x, x+2, color="#1155dd")        #Azulado
ax.plot(x, x+3, color="#15cc55");       #Verdoso
````

#### Estilos de las líneas y los marcadores

La función `plot` tiene diferentes argumentos para definir la forma en que se dibuja la gráfica:
* `linewidth` o `lw`: valor numérico que indica el grosor de la línea.
* `linestyle` o `ls`: cadena de texto que define el estilo de la línea.
* `marker`: texto que indica el marcador que se utiliza para señalar el dato.
* `markersize`: tamaño del marcador
* `markerfacecolor`: color interno del marcador
* `markeredgewidth`: grosor del borde del marcador
* `markeredgecolor`: color del borde del marcador

También se puede definir estilos de línea personalizados con el método `set_dashes` del objeto `line`.

A continuación vamos a ver algunos ejemplos:

````python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

#Posibles estilos de línea: '-', '--', '-.', ':', ...
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')

#Estilo personalizado
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # formato: longitud de la linea, longitud del espacio, ...

#Posibles símbolos marcadores: '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='--', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='--', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='--', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='--', marker='1')

#Tamaño de los marcadores y colores
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
````

### Apariencia de los ejes.

A veces necesitamos cambiar la apariencia de los ejes, indicando donde colocar las etiquetas, modicando el tamaño de las fuentes o definiendo donde empiezan y acaban los ejes.

A continuación vamos a ver como controlar esas propiedades de la figura con Matplotlib.

#### Rango de los ejes.

Lo primero que vamos a configurar es el rango de los ejes. Para ello utilizamos los métodos `set_ylim` y `set_xlim` del objeto `axes`. Para indicar los límites inferior y superior de cada eje utiliza un lista con dos elementos:

````python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("Por defecto")

axes[1].plot(x, x**2, x, x**3)
axes[1].set_ylim([0, 60])
axes[1].set_xlim([2, 5])
axes[1].set_title("Ejes personalizados");
````

#### Escala logarítmica

Otra forma de modificar los ejes es indicando la escala con los métodos `set_xscale` y `set_yscale`. En este caso podemos utilizar el parámetro `"log"` para indicar que la escala es logarítmica:

````python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
````

### Etiquetas y marcas de los ejes.

Podemos indicar explicitamente que marcas queremos poner en cada eje utilizando `set_xticks` y `set_yticks`, las cuales necesitan una lista de valores indicando que marcas queremos mostrar. También podemos utilizar los métodos `set_xticklabels` y `set_yticklabels` para indicar una lista de etiquetas para las marcas del eje:

````python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["${:.1f}$".format(y) for y in yticks], fontsize=18); #etiquetas en estilo LaTeX
````

Existen muchos mas métodos avanzados para controlar el rango de los ejes. Para mas información ver http://matplotlib.org/api/ticker_api.html.

#### Notación científica.

Si en los ejes deben ir números grandes, es mejor utilizar notación científica:

````python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("Notación científica")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
````

### Rejillas

Con el método `grid` del objeto `axes` podemos activar o desactivar una rejilla dentro de la gráfica. También se puede personalizar la apariencia de las líneas de la rejilla usando los mismos argumentos que en el médoto `plot`:

````python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

#Rejilla por defecto
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

#Rejilla personalizada
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
````

También podemos cambiar las propiedades del los bordes de la gráfica.

````python
fig, ax = plt.subplots(figsize=(6,2))

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')

ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

#Quitar borde derecho
ax.spines['right'].set_color("none")
ax.yaxis.tick_left() #Poner marcas sólo en el lado izquierdo
````

### Ejes dobles

A veces es útil tener ejes dobles en `x` o `y`, por ejemplo, cuando dibujamos curvas con diferentes unidades juntas. Matplotlib dispone de los métodos `twinx` y `twiny` en el objeto `axes`:

````python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
#Poner las etiquetas del primer eje en azul
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx() #Creamos el eje gemelo
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
#Poner las etiquetas del segundo eje en rojo
for label in ax2.get_yticklabels():
    label.set_color("red")
````

### Ejes centrados en cero

Si nuestra gráfica tiene valores negativos en x o y, la línea de los ejes aparecerán en los valores mínimos. Si queremos que los ejes aparezca centrados en 0 debemos indicar su posición con `set_position`:

````python
fig, ax = plt.subplots(1,2, figsize=(10,3))
xx = np.linspace(-0.75, 1., 100)

#Ejes por defecto
ax[0].plot(xx, xx**3);

#Ejes centrados en 0
ax[1].spines['right'].set_color('none')
ax[1].spines['top'].set_color('none')

ax[1].xaxis.set_ticks_position('bottom')
ax[1].spines['bottom'].set_position(('data',0)) #Poner el eje x en x=0

ax[1].yaxis.set_ticks_position('left')
ax[1].spines['left'].set_position(('data',0))   #Poner el eje y en y=0

ax[1].plot(xx, xx**3);
````

En Matplotlib existen diversas formas de representar mapas de colores:

````python
#Creamos la matriz de datos que vamos a representar
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
````

# Introducción a Bokeh

Bokeh es una biblioteca de visualización interactiva especialmente diseñada para navegadores web. Entre sus principales características están:

* Visualización interactiva en navegadores modernos.
* Gráficos expresivos y versátiles.

Una de las formas de visualización más comunes es dibujar una gráfica de dispersión de datos usando pequeños glifos *marcadores* para representar cada punto. 

Los pasos a dar son sencillos:
* Crear una figura en blanco, donde dibujar el gráfico.
* Indicar el tipo de glifo que se quiere poner en dicha figura.
* Y utilizar el método `show` para mostrarla.

````python
# Crear una figura o lienzo de 400x400 pixeles
p = figure(width=400, height=400)

# Añadir puntos indicando las coordenadas x e y, junto con otras propiedades (tamaño, color, ...)
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=15, line_color="navy", fill_color="orange", fill_alpha=0.5)

show(p) # Mostrar la gráfica
````

En las gráficas de puntos, el tamaño (`size`) se indica en la misma unidad que la figura (en píxeles), aunque también se puede indicar el radio (`radius`) que utiliza como unidad la escala de los datos.

````python
p = figure(width=400, height=400)
p.square([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=[10, 15, 20, 25, 30], color="firebrick", alpha=0.6)
show(p) 
````

## Imágenes

Otro típo de gráfica común es la que muestra imágenes (por ejemplo mapas de calor, o datos de algún tipo de sensor). Bokeh proporciona dos métodos para mostrar imágenes:

* `image` que puede ser usada, junto con una paleta, para mostrar datos de mapas de color en un gráfico 2D.
* `image_rgba` que puede ser usado para mostrar datos de píxeles RGBA en bruto en una gráfica. 

El siguiente ejemplo muestra cómo utilizar `image` con una matriz 2d y una paleta de color.

````python
N = 500
x = np.linspace(0, 10, N)
y = np.linspace(0, 10, N)
xx, yy = np.meshgrid(x, y) #genera una matriz de coordenadas de 500x500

img = np.sin(xx)*np.cos(yy)

p = figure(x_range=(0, 10), y_range=(0, 10))

# el parámetro 'image' es una matriz de valores
p.image(image=[img], x=0, y=0, dw=10, dh=10, palette="RdYlBu11")

show(p)  
````

Una paleta puede ser cualquier lista de colores. Bokeh ya incorpora paletas de colores predefinidas ([bokeh.palettes](https://bokeh.pydata.org/en/latest/docs/reference/palettes.html)).

```python
r.nonselection_glyph = Circle(fill_alpha=0.2, fill_color="grey", line_color=None) 
```
Cuando se utiliza la interfaz `bokeh.plotting`, es más fácil pasar estas propiedades visuales a los métodos de glifos como se muestra a continuación. El método de glifos creará los glifos de selección o no selección y los adjuntará al renderizador por usted. 

````python
#incluir la herramienta de selección ("tap")
p = figure(width=400, height=400, tools="tap", title="Select a circle")
renderer = p.circle([1, 2, 3, 4, 5], [2, 5, 8, 2, 7], size=50,

                    # Aspectos visuales del los seleccionados.
                    selection_color="firebrick",

                    # Aspectos visuales de los NO seleccionados.
                    nonselection_fill_alpha=0.2,
                    nonselection_fill_color="grey",
                    nonselection_line_color="firebrick",
                    nonselection_line_alpha=1.0)

show(p)
````

## Añadir etiquetas con información a cada dato.

Bokeh tiene diferentes [herramientas](https://bokeh.pydata.org/en/latest/docs/user_guide/tools.html) para sus gráficas, una de ellas es [`HoverTool`](https://bokeh.pydata.org/en/latest/docs/user_guide/tools.html#hovertool). Es particularmente útil para la exploración e interacción de datos. `HoverTool` permite crear cajas de texto que se muestran al pasar el ratón por encima de los puntos de la gráfica. Esta herramienta necesita una lista de tuplas, donde la primera parte de la tupla es un nombre para mostrar y la segunda es un nombre de columna de la fuente de datos precedida por @. Una vez que hemos creado el objeto, lo añadimos al gráfico usando el método `add_tool`. Veamos un ejemplo:

````python
from bokeh.models.tools import HoverTool

# Agrupar los datos por días haciendo la media, máximo y mínimo
days = sea_surface_temperature.index.date
data = sea_surface_temperature.groupby(days).mean() # Se guarda un DataFrame completo
data['max'] = sea_surface_temperature.groupby(days).max()['temperature'] # Añadimos sólo los máximos
data['min'] = sea_surface_temperature.groupby(days).min()['temperature'] # Añadimos sólo los mínimos

data.index.name = 'time' #Ponemos un nombre al índice

pd_source = ColumnDataSource(data)

# Crear gráfica y mostrar
p = figure(x_axis_type="datetime", title="Temperatura del mar por meses", height=350, width=800)
p.xaxis.axis_label = 'Fecha'
p.yaxis.axis_label = 'Valor'

p.line(x='time', y='temperature', source=pd_source)

hover = HoverTool()
hover.tooltips=[
    ('Media', '@temperature'),
    ('Máxima', '@max'),
    ('Minima', '@min')
]

p.add_tools(hover)

show(p)
````
