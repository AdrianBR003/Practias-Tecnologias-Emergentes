#### 28. Considera el vector \[1, 2, 3, 4, 5\], construye un nuevo vector con 3 ceros consecutivos entre cada valor.

````python
Z = np.array([1,2,3,4,5])
nz = 3
Z0 = np.zeros(len(Z) + (len(Z)-1)*(nz))
Z0[::nz+1] = Z
print(Z0)
````

#### 29. Escribe un código que intercambie las dos filas de un array cuadrado de enteros pidiendo por teclado las dimensiones del array los números de fila a intercambiar

````python
dim = int(input("Dimensión de la matriz: "))
M = np.arange(dim*dim).reshape(dim, dim)
print("Original:\n", M)
f1 = int(input("Fila 1 a intercambiar: "))
f2 = int(input("Fila 2 a intercambiar: "))
M[[f1, f2]] = M[[f2, f1]]
print("Resultado:\n", M)
````
