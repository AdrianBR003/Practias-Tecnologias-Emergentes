#### 12. Crea un array de 2 dimensiones relleno de ceros y con el valor 1 en los bordes.

````python
n=np.zeros((10,10))
n[0,:]=1
n[:,0]=1
n[-1,:]=1
n[:,-1]=1
print(n)

````

#### 13. Crea una matriz de 8x8 rellena con un patrón de tablero de ajedrez (utiliza 0 y 1).

````python
n = np.zeros((8,8))
n[1::2, ::2] = 1
n[::2, 1::2] = 1
print(n)
````

````python
np.random.rand(5,5)
min=np.min(n)
max=np.max(n)
n=(n-min)/(max-min)
print(n)
````
